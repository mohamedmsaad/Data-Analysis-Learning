# Analyse des données historiques d'Apple Inc. (AAPL)
This notebook analyzes the historical stock data of Apple Inc. between 2020 and 2025.


# Executive Summary / Résumé Exécutif

This analysis covers:

- Importing AAPL historical data (2020-2025)
- Cleaning and preparing the dataset
- Calculating daily returns
- Computing moving averages
- Assessing rolling volatility
- Visualizing trends and risk metrics

This project demonstrates fundamental data analysis techniques applied to financial time series.

---


## Chargement des données
We load the CSV file exported from Nasdaq.


```python
import pandas as pd

data = pd.read_csv("AAPL.csv")
print(data.columns)
print(data.head())
```

## Nettoyage des données
We check for missing values and prepare the dataset.

```python
# Nettoyage des colonnes
data.columns = data.columns.str.strip()

# Conversion brute
data['Close/Last_clean'] = (
    data['Close/Last']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.strip()
    .astype(float)
)
```

## Visualisation générale
Graphs of price trends and returns distributions.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Activate seaborn styling
sns.set()


plt.figure(figsize=(16,8))
plt.plot(pd.to_datetime(data['Date']), data['Close/Last_clean'])
plt.title("AAPL Close Price")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Calcul des rendements journaliers
We calculate daily returns to evaluate volatility.


```python
# Calculate daily returns
data['Daily Return'] = data['Close/Last_clean'].pct_change()

# Plot daily returns
plt.figure(figsize=(12,6))
plt.plot(data['Daily Return'])
plt.title("AAPL Daily Returns")
plt.xlabel("Date")
plt.ylabel("Return")
plt.grid(True)
plt.show()
```

## Visualisation des tendances
Graphs of price trends and returns distributions.


```python
# Histogram of daily returns
plt.figure(figsize=(10,5))
data['Daily Return'].hist(bins=50)
plt.title("Histogram of AAPL Daily Returns")
plt.xlabel("Return")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

## Statistiques descriptives
Overview of key metrics.

```python
# Summary statistics
print(data['Daily Return'].describe())
```

## Moving averages

Affichage des moving averages sur 20 et 50 jours.

```python
# Compute moving averages
data['20D MA'] = data['Close/Last_clean'].rolling(window=20).mean()
data['50D MA'] = data['Close/Last_clean'].rolling(window=50).mean()

# Plot with moving averages
plt.figure(figsize=(12,6))
plt.plot(data['Close/Last_clean'], label='Close')
plt.plot(data['20D MA'], label='20-Day MA')
plt.plot(data['50D MA'], label='50-Day MA')
plt.title("AAPL Closing Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)
plt.show()
```

## Analyse de la volatilité / Volatility Analysis

We calculate the 20-day rolling standard deviation of daily returns to assess periods of higher risk.


```python
# Calculate rolling volatility (standard deviation of daily returns)
data['Volatility_20D'] = data['Daily Return'].rolling(window=20).std()

# Plot volatility
plt.figure(figsize=(12,6))
plt.plot(data['Volatility_20D'], label='20-Day Rolling Volatility')
plt.title('AAPL 20-Day Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.show()
```
